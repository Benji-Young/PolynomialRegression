```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from sklearn.metrics import r2_score

# Generate some sample data
np.random.seed(0)
n = 100
x = (np.random.rand(n) * 100) - 50
y = x**3 + 2 * x**2 + 3 * x + 1 + (np.random.rand(n) * 100) - 50

# Create a list of degrees to test
degrees = [2, 3, 4, 5, 6]

# Create a list to store the R^2 scores for each degree
r2_scores = []

# Loop over the degrees
for degree in degrees:
    # Create Polynomial Features
    poly_features = PolynomialFeatures(degree=degree)
    x_poly = poly_features.fit_transform(x.reshape(-1, 1))

    # Create Linear Regression Object
    lin_reg = LinearRegression()

    # Train the model
    lin_reg.fit(x_poly, y)

    # Make predictions
    y_pred = lin_reg.predict(x_poly)

    # Calculate R^2 score
    r2_score_value = r2_score(y, y_pred)

    # Append the R^2 score to the list
    r2_scores.append(r2_score_value)
```

```python
# Plot the R^2 scores
plt.plot(degrees, r2_scores)
plt.xlabel('Degree')
plt.ylabel('R^2 Score')
plt.title('Polynomial Regression (Degree vs. R^2 Score)')
plt.show()
```
